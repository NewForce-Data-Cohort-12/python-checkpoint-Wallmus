### 3. Using movies_df, subset the data to create a new dataframe named year_duration that contains only two columns: release_year and duration


```python
#new dataframe with just the two colums, though it still includes tv shows
year_duration = movie_disneyplus[['title','duration','rating','release_year','date_added']]
year_duration.shape
year_duration.head()
```

### 4. Using your year_duration dataframe, filter for movies released between 1975 and 1984 (inclusive) and find the minimum duration, maximum duration

```python
# only release_year from 1975 through 1984 please
filtered_years = year_duration[(year_duration['release_year'] >= 1975) & (year_duration['release_year'] <= 1984)]

# extract first group of digits from duration, convert to int, and save in new column duration_int
filtered_years['duration_int'] = filtered_years['duration'].str.extract('(\d+)').astype(int)
year_duration['duration_int'] = year_duration['duration'].str.extract('(\d+)').astype(int)

# filter out rows with "Season" in the duration column because they're tv obviously
# between '75 and '84
movies_filtered = filtered_years[~filtered_years['duration'].str.contains('Season', na=False)]

# another dataframe without the year constraint for later
movies_unfiltered = year_duration[~year_duration['duration'].str.contains('Season', na=False)]

# print min, max, and average
print("\nMinimum duration:", movies_filtered['duration_int'].min())
print("Maximum duration:", movies_filtered['duration_int'].max())
print("Average duration:", movies_filtered['duration_int'].mean(),'\n')

#how many movies are in here?
print('1975-1984:',movies_filtered.shape, '\nall movies:',movies_unfiltered.shape)
```

### 5. Using the year_duration dataframe, create a scatterplot that shows movie durations over the years. Add a descriptive title to your chart.

```python
#plot thickens
plt.scatter(movies_unfiltered['release_year'], movies_unfiltered['duration_int'])
plt.title('Movie Durations Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Duration (min)')
plt.show()
```

### Do you notice any correlation between release year and movie duration?

Movie duration appears to trend upward as the year increases, and the standard deviation increases dramatically.

### 6. Find the actual correlation value between release_year and duration.

```python
#used movies_filtered instead of year_duration to avoid including tv shows
year_corr_duration = movies_unfiltered[['release_year', 'duration_int']]
print(year_corr_duration.corr())
```

### Based on the correlation value, is there a meaningful correlation between release year and duration?

At 0.24, we are seeing a positive correlation. This means that though there is variation, the average movie duration is going up over the years.

##### 2. Create a new dataframe containing only movies from that year.

```python
covid_movies = movies_unfiltered[movies_unfiltered['release_year'] == 2020]
covid_movies.head()
```

##### 3. Convert the duration column to integers.

```python
#whoops kinda did this earlier in #4 with the duration_int column
covid_movies['duration'] = covid_movies['duration_int']
del covid_movies['duration_int']
covid_movies.head()
```

##### 4. Use the IQR method to determine the upper and lower cutoffs for outliers.

```python
# find quartiles from duration
Q1 = covid_movies['duration'].quantile(0.25)
Q3 = covid_movies['duration'].quantile(0.75)

#calculate IQR
IQR = Q3 - Q1

# calculate adn print lower and upper outlier cutoffs
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR
print("Lower cutoff:",lower)
print("Upper cutoff:",upper)
```

##### 5. Calculate what percentage of movies from that year had durations that were statistical outliers.

```python
# new dataframe with durations between the upper and lower outlier cutoffs
outlier_movies = covid_movies[(covid_movies['duration'] < lower) | (covid_movies['duration'] > upper)]
outlier_percentage = len(outlier_movies) / len(covid_movies) * 100
print(round(outlier_percentage, 2), '% of movies are outliers')
```

### 9. Add a new column called length_category to movies_df that categorizes each movie as one of the following

```python
# define function that returns one of three strings based on the duration length
def len_cat(duration):
    if duration < 30:
        return 'short'
    elif duration < 120:
        return 'average'
    else:
        return 'long'

#use our new len_cat function in the apply method with a row's duration_int value and store it in new column length_category
movies_unfiltered['length_category'] = movies_unfiltered['duration_int'].apply(len_cat)
movies_unfiltered.head()
```
